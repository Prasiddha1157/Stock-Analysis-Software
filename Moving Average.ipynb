```python
yf.pdr_override()
stock = input("Enter a stock ticker symbol: ")
print(stock)
```

```python
startyear = 2019
startmonth = 1
startday = 2

start = dt.datetime(startyear, startmonth, startday)
now=dt.datetime.now()

df = pdr.get_data_yahoo(stock, start, now)
```

```python
ma = 50 # moving average

smaString = 'Sma_' + str(ma) # simple moving average

df[smaString] = df.iloc[:, 4].rolling(window = ma).mean() # window means the size of the moving window and rolling is for rolling window calculations and 4 because moving average calculated using Adj Close 
```

```python
df = df.iloc[ma:]
df
```

```python
numH = 0
numC = 0

for i in df.index:
    #print(df[smaString][i]) # print(df.iloc[:, 4][i])
    if (df['Adj Close'][i] > df[smaString][i]):
        print("The close is higher.")
        numH += 1
    else:
        print("The close is lower.")
        numC += 1
    
print("Close was higher " + str(numH) + " times.")
print("Close was lower " + str(numC) + " times.")
```
